### Hopefully you know that each `tr` is supposed to be a row of your data. What is the index of the first row element that is actually a result?

`.text` will help you here.

```python
table = driver.find_element_by_xpath('//*[@id="content"]/table[3]')
```

```python
first_row = table.find_elements_by_tag_name('tr')[1]
first_row.text
```

### Loop through each operator result, printing its name

You can use list slicing or an `if` statement to skip the non-data row(s).

```python
# operater name is always the third 'td'
rows = table.find_elements_by_tag_name('tr')
for row in rows[1:-1]:
    operator_name = row.find_elements_by_tag_name('td')[2]
    print(operator_name.text)
```

### Loop through each operator result, printing its ID

There should be ONE code per row, and NO empty rows between them.

```python
rows = table.find_elements_by_tag_name('tr')
for row in rows[1:-1]:
    id_number = row.find_elements_by_tag_name('td')[0]
    print(id_number.text)
```

## Saving the results

### Loop through each `tr` to create a list of dictionaries

Each dictionary must contain

- Operator ID
- Operator name
- Mine name
- State
- Mine type
- Coal or metal
- Status
- Commodity

Create a new dictionary for each row.

```python
mylist = []
rows = table.find_elements_by_tag_name('tr')
for row in rows[1:-1]:
    dic ={}
    
    id_number = row.find_elements_by_tag_name('td')[0]
    dic['Operator ID'] = id_number.text
    
    operator_name = row.find_elements_by_tag_name('td')[2]
    dic['Operator name'] =operator_name.text
    
    mine_name = row.find_elements_by_tag_name('td')[3]
    dic['Mine name'] =mine_name.text
    
    state = row.find_elements_by_tag_name('td')[1]
    dic['State']=state.text
    
    mine_type = row.find_elements_by_tag_name('td')[4]
    dic['Mine type']=mine_type.text
    
    coal_metal = row.find_elements_by_tag_name('td')[5]
    dic['Coal or metal']=coal_metal.text
    
    commodity = row.find_elements_by_tag_name('td')[6]
    dic['Commodity']= commodity.text
    
    mylist.append(dic)
mylist
```

### Save that to a CSV named `dirt-operators.csv`

```python
import pandas as pd
df =pd.DataFrame(mylist)
df.head()
```

```python
df.to_csv('dirt-operators.csv',index=False)
```
